```
import pandas as pd
import numpy as np
import json
df = pd.read_json(path+'test_pred_e15.json')
df
```

```
#back-process the data into slot!
print(df["text"].iloc[0])
print(len(df["text"].iloc[0].split()))
print(len(df["pred_tags"].iloc[0]))
print(df["pred_tags"].iloc[0][:8]) #num of length
```

```
#add new sent_length column
sent_len = [] #len = 2887
for sent in df["text"]:
  #print(sent)
  sent_len.append(len(sent.split()))
print(sent_len)
df["sent_length"] = sent_len
df
```

```
#clean pred_tag into correct sentence length
#df["pred_tags"].iloc[0][:8]

i = 0
for num in sent_len:
  df["pred_tags"].iloc[i] = df["pred_tags"].iloc[i][:int(num)]
  i += 1

df
```

```
# add intent list 
df_in = pd.read_json(path+'test_pred_e5.json')
#df_in = df
tags_for_intent = []
#delete "O"
for tag in df_in["pred_tags"]:
  tag = list(set(tag))
  if "O" in tag and len(tag) > 1:
    tag.remove("O")
  tags_for_intent.append(tag)
df_in["pred_tags"] = tags_for_intent

#create intent
intent = []
#['AddToPlaylist', 'BookRestaurant', 'GetWeather', 'PlayMusic', 'RateBook', 'SearchCreativeWork', 'SearchScreeningEvent']

for i in range(1439):
  #print(i)
  #print(df["pred_tags"].iloc[i])
  #print(df["pred_tags"].iloc[i][0])
  if "ATP" in df_in["pred_tags"].iloc[i][0]:
    intent.append("AddToPlaylist")
  if "BR" in df_in["pred_tags"].iloc[i][0]:
    intent.append("BookRestaurant")
  if "GW" in df_in["pred_tags"].iloc[i][0]:
    intent.append("GetWeather")
  if "PM" in df_in["pred_tags"].iloc[i][0]:
    intent.append("PlayMusic")
  if "RB" in df_in["pred_tags"].iloc[i][0]:
    intent.append("RateBook")
  if "SCW" in df_in["pred_tags"].iloc[i][0]:
    intent.append("SearchCreativeWork")
  if "SSE" in df_in["pred_tags"].iloc[i][0]:
    intent.append("SearchScreeningEvent")
  if "O" in df_in["pred_tags"].iloc[i][0]:
    intent.append("AddToPlaylist")

print(len(intent))
df_in["intent"] = intent
df_in
```

```
#add slot dictionary

ATP_slot_keys = ['music_item', 'playlist_owner', 'playlist', 'entity_name', 'artist']
BR_slot_keys = ['restaurant_type', 'city', 'party_size_number', 'cuisine', 'timeRange', 'state', 'served_dish', 'restaurant_name', 'spatial_relation', 'poi', 'country', 'party_size_description', 'facility', 'sort']
GW_slot_keys = ['country', 'timeRange', 'city', 'state', 'condition_temperature', 'current_location', 'spatial_relation', 'condition_description', 'geographic_poi']
PM_slot_keys = ['music_item', 'album', 'artist', 'track', 'service', 'sort', 'year', 'genre', 'playlist']
RB_slot_keys = ['object_select', 'object_type', 'rating_value', 'rating_unit', 'best_rating', 'object_name', 'object_part_of_series_type']
SCW_slot_keys = ['object_name', 'object_type']
SSE_slot_keys = ['location_name', 'movie_type', 'object_type', 'movie_name', 'spatial_relation', 'object_location_type', 'timeRange']

ATP_slot_dict = {}
for tag in ATP_slot_keys:
  ATP_slot_dict[tag] = ["B_ATP_" + tag, "I_ATP_" + tag]
print("ATP_slot_dict: ", ATP_slot_dict)

BR_slot_dict = {}
for tag in BR_slot_keys:
  BR_slot_dict[tag] = ["B_BR_" + tag, "I_BR_" + tag]
print("BR_slot_dict: ", BR_slot_dict)

GW_slot_dict = {}
for tag in GW_slot_keys:
  GW_slot_dict[tag] = ["B_GW_" + tag, "I_GW_" + tag]
print("GW_slot_dict: ", GW_slot_dict)

PM_slot_dict = {}
for tag in PM_slot_keys:
  PM_slot_dict[tag] = ["B_PM_" + tag, "I_PM_" + tag]
print("PM_slot_dict: ", PM_slot_dict)

RB_slot_dict = {}
for tag in RB_slot_keys:
  RB_slot_dict[tag] = ["B_RB_" + tag, "I_RB_" + tag]
print("RB_slot_dict: ", RB_slot_dict)

SCW_slot_dict = {}
for tag in SCW_slot_keys:
  SCW_slot_dict[tag] = ["B_SCW_" + tag, "I_SCW_" + tag]
print("SCW_slot_dict: ", SCW_slot_dict)

SSE_slot_dict = {}
for tag in SSE_slot_keys:
  SSE_slot_dict[tag] = ["B_SSE_" + tag, "I_SSE_" + tag]
print("SSE_slot_dict: ", SSE_slot_dict)
```

```
slot = []
slot_dict = {}

for i in range(1439): #2887
  keys = df["pred_tags"].iloc[i]
  values = df["text"].iloc[i].split()
  for j in range(len(keys)):
    #print("keys[j]: ", keys[j]) #'B_ATP_entity_name' etc
    #print("values[j]: ", values[j]) #'changes' etc
    for k, v in ATP_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in BR_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in GW_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in PM_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in RB_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in SCW_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])
    for k, v in SSE_slot_dict.items(): #v = ['B_ATP_music_item', 'I_ATP_music_item'] etc
      if (keys[j] in v) and (k not in slot_dict):
        slot_dict[k] = str(values[j])
      if (keys[j] in v) and (k in slot_dict) and (slot_dict[k] != str(values[j])) :
        slot_dict[k] += " " + str(values[j])

  slot.append(slot_dict)
  slot_dict = {}

#print(slot)
print(len(slot))
df["intent"] = df_in["intent"]
df["slots"] = slot
df
```
